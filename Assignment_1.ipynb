Q2. What is the difference between string and variable?

Variable: A variable is a programming element used to define, store and perform operations. A variable can either be an integer, string, float, etc.

String: A string is a type of a variable. The value in a string variable is stored between quotes, either single or double. It is a data type for a sequence of characters. 

Q3. Describe three different data types.

The 3 main data types are:
1. Numeric: Data that has numeric value. Numeric value can be integer, float or complex number.

2. Sequence : Sequence is the ordered collection of similar or different data types. Sequence allows to store multiple values in an organized and efficient fashion. Sequence types are: string, list, tuple

3. Boolean: Boolean objects are two types, True or False.

Two other data types are:
1. Set: It is an unordered collection of data type that is iterable, mutable and has no duplicate elements.

2. Dictionary: It is an unordered collection of data values. Dictionay holds key:value pair.

Q5. The assignment statements, like spam=10. What is the difference between and expression and statement?


Expression: An expression is a combination of symbols, like variables, constants, operators and calls to functions. It represents a value and is evaluated to produce the value. An expression must have atleast one operand (constant or variable) and have one or more operators. e.g x+7-5, "a"+"b"

Statement: A statement does something but does not return a value. eg. assignment statements, conditional branching, loops, classes, import, etc.

Q10. Why does this expression cause an error? How can you fix it?
'I have eaten'+99+'burritos.'

This expression causes an error because the addition operation is between string and integer values. The type of values should be same.

It can be fixed by converting all the values to string.

'I have eaten'+'99'+'burritos.'

This will run to give: 'I have eaten 99 burritos.'
